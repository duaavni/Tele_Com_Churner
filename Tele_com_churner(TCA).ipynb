```python
# Normalize the data to percentages
df_grouped = df.groupby(['Senior Citizen', 'Churn Label']).size().unstack().fillna(0)
df_grouped = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

# Plotting the stacked bar chart
ax = df_grouped.plot(kind='bar', stacked=True, figsize=(5,5), colormap='viridis')

# Add labels and title
plt.title("Churn by Senior Citizen (Percentage of Total)")
plt.ylabel("Percentage (%)")
plt.xlabel("Senior Citizen")

# Adding percentage labels on top of each bar segment
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # Avoid adding labels for zero heights
        ax.text(p.get_x() + p.get_width() / 2., p.get_height() + p.get_y(), 
                f'{height:.1f}%', 
                ha='center', va='bottom', fontsize=9, color='black')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selecting categorical columns (excluding numerical and ID columns)
categorical_columns = [
    'Gender', 'Senior Citizen', 'Partner', 'Dependents', 'Phone Service', 
    'Multiple Lines', 'Internet Service', 'Online Security', 'Online Backup', 
    'Device Protection', 'Tech Support', 'Streaming TV', 'Streaming Movies', 
    'Contract', 'Paperless Billing', 'Payment Method'
]

# Setting up the figure size and subplots
num_cols = 3  # Number of columns in subplot grid
num_rows = -(-len(categorical_columns) // num_cols)  # Ceiling division for rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 4))
axes = axes.flatten()  # Flatten to easily iterate

# Loop through each categorical column and create a count plot
for i, col in enumerate(categorical_columns):
    sns.countplot(x=df[col], hue=df["Churn Label"], ax=axes[i], palette="viridis")
    axes[i].set_title(f'Distribution of {col} by Churn Label')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate labels for better readability

# Remove empty subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
#some problem with electronic check
#people who took month-to-month plan churn a lot
#those who have not taken online back-up churn a lot
#also people who have not taken device protection, online security,tech support they also churn a lot
#people who took optic fibre are comparatively churning more
```

Analysis of Churn Data (Based on the Plots)

1. Contract & Internet Service Influence Churn → Customers with month-to-month contracts and those with fiber optic internet show higher churn rates, likely due to flexibility and potential service dissatisfaction.  
2. Tech Support & Online Services Matter → Customers without tech support, online security, or backup services tend to churn more, indicating that better service engagement reduces churn.  
3. **Senior Citizens & Multiple Lines Impact → Senior citizens and customers with multiple lines have a higher churn rate, suggesting potential issues with affordability or service complexity.  
4. Payment Method & Billing Affect Churn → Electronic check users and those with paperless billing show increased churn, possibly due to unexpected charges or ease of switching providers.
